```python

import cv2
import numpy
import os

# Make an array of 120,000 random bytes.
randomByteArray = bytearray(os.urandom(120000))
print(randomByteArray)
print("------------------")

flatNumpyArray = numpy.array(randomByteArray)
print(flatNumpyArray)
print("------------------")

# Convert the array to make a 400x300 grayscale image.
grayImage = flatNumpyArray.reshape(300, 400)
cv2.imwrite('sources/RandomGray.png', grayImage)
print(grayImage)
print("------------------")

# Convert the array to make a 400x100 color image.
bgrImage = flatNumpyArray.reshape(100, 400, 3)
cv2.imwrite('sources/RandomColor.png', bgrImage)

```

# conocer los tamaños de una imagen

* shape: This is a tuple describing the shape of the array. For an image, it contains (in order) the height, width, and—if the image is in color—the number of channels. The length of the shape tuple is a useful way to determine whether an image is grayscale or color. For a grayscale image, we have len(shape) == 2, and for a color image, len(shape) == 3.

* size: This is the number of elements in the array. In the case of a grayscale image, this is the same as the number of pixels. In the case of a BGR image, it is three times the number of pixels because each pixel is represented by three elements (B, G, and R).

* dtype: This is the datatype of the array's elements. For an 8-bit-per-channel image, the datatype is numpy.uint8.



```python
##Conocer los tamaños de una imagen
import cv2
img = cv2.imread('sources/goku.png')
print(img.shape)
print(img.size)
print(img.dtype)
```

# Converting between numpy.array and cv2.UMat

Currently, cv2.UMat supports GPU acceleration on systems that have OpenCL-compliant GPU drivers, with a minimum version of OpenCL 1.2. Most modern desktop or laptop GPUs, as well as some high-end smartphone or tablet GPUs, have this level of OpenCL support.


The following code initializes a UMat by copying data from a NumPy array:


`my_umat = cv2.UMat(my_numpy_array)`


Most OpenCV functions will accept a UMat argument in place of NumPy array and, then, will return a UMat instead of NumPy array. Here is an example:


`my_converted_umat = cv2.cvtColor(my_umat, cv2.COLOR_GRAY2BGR)`

NumPy functions will not accept a UMat argument, a UMat does not have the same methods as a numpy.array, and we cannot perform slicing or other Pythonic operations on UMat data. However, we can convert a UMat back to a NumPy array, as follows:

`my_converted_numpy_array = my_converted_umat.get()`

Tal vez se pregunte, ¿debería comenzar a usar s en todas partes para acelerar sus aplicaciones? Probablemente no. Copiar datos dentro y fuera de la memoria de la GPU es una operación costosa y, en muchos casos, este costo general supera el beneficio de un mayor paralelismo en la GPU. Por otro lado, si puede minimizar el número de copias y si puede aplicar una serie de funciones OpenCV a una gran cantidad de datos en la GPU, entonces el beneficio puede superar el costo.UMat

A lo largo de este libro, usaremos principalmente en lugar de , pero tenga en cuenta que es posible que desee probar el rendimiento de como alternativa para sus propias aplicaciones prácticas.

```python
import cv2
import numpy as np

img = cv2.imread('sources/goku.png', cv2.IMREAD_GRAYSCALE)

my_umat = cv2.UMat(img)


cv2.imshow("imagen",my_umat)
k = cv2.waitKey(0)
cv2.destroyAllWindows()


nuevoNumpyImagen = my_umat.get()
cv2.imshow("imagen",nuevoNumpyImagen)
k = cv2.waitKey(0)
cv2.destroyAllWindows()



print(my_umat)
print(nuevoNumpyImagen)
```
